# Shannon Entropy Feature Extraction

```python
import pandas as pd
import numpy as np
from scipy.stats import entropy
```

```python
df = pd.read_csv('feature_files/normalized_merged_normal.csv', keep_default_na=False, na_values=[""])
df
```

```python
table = {}
for i in range(7):
    col = str(i)
    N = len(df[col])
    entropy_values = []
    for j in range(0, N, 120):
        pd_series = pd.Series(df[col][j:j + 120])
        counts = pd_series.value_counts()
        val = entropy(counts)
        entropy_values.append(val)
        
    table[col] = entropy_values
print(table)
```

```python
result = pd.DataFrame(table)
result.to_csv('feature_files/entropy_normalized_merged_normal.csv')
```
