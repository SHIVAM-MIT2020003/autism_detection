```python
import numpy as np
import mne
import os
from os import listdir
from os.path import isfile, join
mypath = 'data/normal/'
files = [f for f in listdir(mypath) if isfile(join(mypath, f))]

n = len(files)
# print(files)
raws = []

for i in range(n):
    raws.append(mne.io.read_raw_edf('data/normal/{}'.format(files[i])))
    
result = mne.concatenate_raws(raws)
result.save('new_raw.fif')


```

```python
raw = mne.io.read_raw_fif('feature_files/merged_normal_raw.fif')
```

```python
raw.plot()
```

```python
import numpy as np
import mne
import os
from os import listdir
from os.path import isfile, join

mypath = 'data/autism/'
files = [f for f in listdir(mypath) if isfile(join(mypath, f))]

n = len(files)
print(files)

raws = []

for i in range(n):
    raws.append(mne.io.read_raw_edf('data/autism/{}'.format(files[i])))
    
result = mne.concatenate_raws(raws, preload=True)
print(result)
result.save('autism-raw.fif')
```

```python
raw = mne.io.read_raw_fif('feature_files/merged_normal_raw.fif')
raw.plot()
```
