# Skewness feature extraction

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('feature_files/normalized_merged_normal.csv', keep_default_na=False, na_values=[""])
df
```

```python
table = {}
for i in range(7):
    col = str(i)
    N = len(df[col])
    var = []
    for j in range(0, N, 120):
        var.append(df[col][j:j + 120].skew(axis = 0, skipna = True))
    table[col] = var
print(table)
```

```python
result = pd.DataFrame(table)
result.to_csv('feature_files/skewness_normalized_merged_normal.csv')
```
