# Segmentation

```python
import  mne
import matplotlib.pyplot
file = "feature_files/merged-normal-raw.fif"

raw = mne.io.read_raw_fif(file, preload=True)

montage = mne.channels.make_standard_montage('standard_1020')

raw.set_montage(montage, on_missing='ignore')
```

```python
events = mne.find_events(raw)
```

```python
mne.viz.plot_events(events[:4])
```

# mne.Epochs function is used for segmentation

```python
event_ids = {"standard/stimulus":4, "target/stimulus": 1}
epochs = mne.Epochs(raw, events, event_id=event_ids, preload = True)
```

# Segmented DataSet

```python
epochs.plot()
```

```python
epochs["target"]
```

```python
epochs.info
```

```python
for i in range(7):
    epochs["target"].plot_image(picks=[i])
```

```python
for i in range(7):
    epochs["standard"].plot_image(picks=[i])
```

```python
epochs.equalize_event_counts(event_ids)
```

```python
epochs.save("segmented-merged-normal-epo.fif")
```
