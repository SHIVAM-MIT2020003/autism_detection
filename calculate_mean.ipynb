# Calculate Mean for each Segment

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('feature_files/normalized_merged_autism.csv', keep_default_na=False, na_values=[""])
df
```

```python
# df['0']
# df['1']
# len(df['1'])
# print(df['1'][10:20])
```

```python
map = {}
for i in range(7):
    col = str(i)
    N = len(df[col])
    mean = []
    for j in range(0, N, 120):
        mean.append(df[col][j:j + 120].mean())
    map[col] = mean
```

```python

# print(map['0'])
```

```python
result = pd.DataFrame(map)
result.to_csv('feature_files/mean_normalized_merged_autism.csv')
```
